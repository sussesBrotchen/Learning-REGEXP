## Задачки с использованием REGEX

### Изученные функции:
- re.search
- re.sub
- re.findall

```python
# delete symbols from text using regexp

import re


text = '''While Python remains a dynamically typed language, meaning you don't have to specify types for variables, incorporating type annotations, especially in larger projects or when working in a team, can significantly enhance code quality and maintainability.
It's worth noting that type annotations are optional, and Python will not enforce strict typing; they serve primarily as a development aid and documentation tool.'''

print(re.sub(r'[^\w\s]', '', text))
```

```python
# find first number


import re

text = 'Мне 25 лет, а моей сестре 30'
match = (re.search(r'\d+', text))

if match:
    print('Нашёл:', match.group())  # group() возвращает то, что нашлось
else:
    print('Ничего не нашёл')
```

```python
# is there an e-mail in the text


import re

text = 'Мои контакты: john@mail.com и anna@gmail.com'
match = (re.search(r'[\w]+[@][\w]+[\.][\w]+', text))

if match:
    print('Нашёл:', match.group())  # group() возвращает то, что нашлось
else:
    print('Ничего не нашёл')
```

```python
# Find the first word with capital


import re

text = 'маша купила Молоко и Хлеб. потом встретила Диму'
match = (re.search(r'[А-Я][а-я]+', text))

if match:
    print('Нашёл:', match.group())  # group() возвращает то, что нашлось
else:
    print('Ничего не нашёл')
```

```python
# Find the first word with capital


import re

text = 'маша купила Молоко и Хлеб. потом встретила Диму'
matches = (re.findall(r'[А-Я][а-я]+', text))

if matches:
    print('Нашёл:', matches)  # group() возвращает то, что нашлось
else:
    print('Ничего не нашёл')
```

```python
# replace phone number with word


import re

text = 'Позвони мне по номеру +7 999 123-45-67 после шести'
match = re.sub(r'([\+]?[7]?[ ]?[\d]{3}[ ]?[\d]{3}[ -]?[\d]{2}[ -]?[\d]{2})', '[телефон]', text)

print(match)
```

```python
# print a date if it is in data


import re

text = 'Встреча назначена на 15.05.2023'
match = (re.search(r'[\d]{2,4}[\.][\-]?[\d]{2}[\.][\-]?[\d]{2,4}', text))

if match:
    print('Нашёл:', match.group())  # group() возвращает то, что нашлось
else:
    print('Ничего не нашёл')
```

```python
# print dates if it is in data


import re

text = 'Даты: 1.05.2023, 15.5.2023, 15.05.23, 2023.05.15'
matches = (re.findall(r'[\d]{1,4}[\.][\-]?[\d]{1,2}[\.][\-]?[\d]{1,4}', text))

if matches:
    print('Нашёл:', matches)  # matches — это список
else:
    print('Ничего не нашёл')
```
